## Группировка категориальных переменных

```python
df.groupby('education_level',dropna=False)['target'].agg(['count','mean'])
```

Объединим пары групп

Строки не перебераем циклом(!!!!), столбцы иногда можно

```python
def ed_group(x): #каждое значение из education_level
    if x in ['High School', 'Primary School']:
        return 'School'
    if x in ['Masters', 'Phd']:
        return 'Masters_phd'
    return x
```

```python
df['new_education_level'] =  df['education_level'].apply(ed_group) #с помощью apply все делается
```

```python
df.groupby('new_education_level',dropna=False)['target'].agg(['count','mean']) #сводники
```

ЕСТЬ второй способ сделать это:

```python
df['education_level'].replace('Primary School','School')
```

PHD добавим количество часов

```python
def new_training_hours(row):
    if row['education_level'] == 'Phd':
        return row['training_hours'] + 1000
    return row['training_hours']
```
